```python
import numpy as np
import matplotlib.pyplot as plt
```

### Original Image

```python
Image = np.array(plt.imread("/content/drive/MyDrive/Image collection/robo_dalle.jpg"))
plt.imshow(Image)
```

**Red, Green and Blue channel**



```python
Image_r  = Image.copy(); Image_g  = Image.copy(); Image_b  = Image.copy();
Image_r[:,:,(1,2)]=0; Image_g[:,:,(0,2)]=0; Image_b[:,:,(0,1)]=0
plt.imshow(np.hstack((Image_r,Image_g,Image_b)))
```

**Gray version of red, green and blue**

```python
plt.imshow(np.hstack((Image_r[:,:,0],Image_g[:,:,1],Image_b[:,:,2])),cmap="gray")
```

**Rotate left, right and upside down**

```python
Image_left = Image.copy(); Image_right = Image.copy(); Image_down = Image.copy()
Image_left = np.rot90(Image_left,k=1); Image_right = np.rot90(Image_right,k=-1); Image_down = np.rot90(Image_down,k=-2)
plt.imshow(np.hstack((Image_left,Image_right,Image_down)))
```

**Negative Image**

```python
Image1 = Image.copy()
Image_negative = 255 - Image1
plt.imshow(Image_negative)
```

**Cropped Image**

```python
Image2 = Image.copy()
Image_crop = Image2[200:501,200:801,:]
plt.imshow(Image_crop)
```

**Mirror Image**

```python
Image3 = Image.copy()
Image_mirror = Image3[:,::-1,:]
plt.imshow(np.hstack((Image,Image_mirror)))
```

**Water Image**

```python
Image4 = Image.copy()
Image_water = Image[::-1,:,:]
plt.imshow(np.vstack((Image,Image_water)))
```
