```python
import matplotlib.pyplot as plt
```

```python
Dot_image = np.array(plt.imread("/content/drive/MyDrive/Colab Notebooks/Dot_image.png"))
plt.imshow(Dot_image)
```

```python
print(Dot_image)
```

```python
print(Dot_image.shape);print(Dot_image.ndim)
```

```python
Image_r = Dot_image.copy()
Image_r[:,:,[1,2]]=0
plt.imshow(Image_r)
```

```python
Image_r_gray = Image_r[:,:,0]
plt.imshow(Image_r[:,:,0], cmap="gray")
```

```python
Image_g = Dot_image.copy()
Image_g[:,:,[0,2]]=0
plt.imshow(Image_g)
```

```python

plt.imshow(Image_g[:,:,1],cmap="gray")
```

```python
Image_b = Dot_image.copy()
Image_b[:,:,(0,1)]=0
plt.imshow(Image_b)
```

```python
plt.imshow(Image_b[:,:,2], cmap = "gray")
```

```python
Image_rgb = np.hstack((Image_r,Image_g,Image_b))
plt.imshow(Image_rgb)
```

```python
Image_rgb_gray = np.hstack((Image_r[:,:,0],Image_g[:,:,1],Image_b[:,:,2]))
plt.imshow(Image_rgb_gray,cmap="gray")
```
